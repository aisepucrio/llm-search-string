```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from io import StringIO
```

# Boxplot F1-Score Scopus

```python
df_scopus = pd.read_csv("results2/metrics_scopus/RESULTADOS.csv")

# Remover sufixo ':latest' dos modelos para padronizar
df_scopus['modelo'] = df_scopus['modelo'].str.replace(':latest', '', regex=False)

# Selecionar apenas colunas de interesse
df_filtered_scopus = df_scopus[['modelo', 'prompt_name', 'F1 Score']].copy()

# Normalizar F1 Score (entre 0 e 1)
scaler = MinMaxScaler()
df_filtered_scopus['F1 Score Normalizado'] = scaler.fit_transform(df_filtered_scopus[['F1 Score']])

# Plot dos boxplots normalizados
plt.figure(figsize=(8, 4))
sns.boxplot(data=df_filtered_scopus, x='modelo', y='F1 Score Normalizado', hue='prompt_name')
plt.title('Normalized Distribution of F1 Score by Modelo and Prompt')
plt.xlabel('Modelo')
plt.ylabel('F1 Score (Normalized)')
plt.legend(title='Prompt')
plt.grid(True)
plt.ylim(0, 1.4)
plt.tight_layout()
plt.show()
```
